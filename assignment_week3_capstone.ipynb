```python
import numpy as np
import pandas as pd
import requests

website_url = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
from bs4 import BeautifulSoup
soup = BeautifulSoup(website_url,'lxml')

print(soup.prettify()) # where we can see html code and find what we want
print(soup.title.text)
My_table = soup.find('table', class_='wikitable')
My_table

t_headers = []
for th in My_table.find_all('th'):
    t_headers.append(th.text.replace('/n','').strip())
    
table_data =[]

for tr in My_table.find_all('tr'):
    t_row = {}
    for td,th in zip(tr.find_all('td'),t_headers):
        t_row[th] = td.text.replace('/n','').strip()
    table_data.append(t_row)
               
print(table_data)
import csv

with open(f'Toronto_neighborhood.csv','w') as out_file:
    headers = ["Postal code","Borough","Neighborhood"]
    writer = csv.DictWriter(out_file,headers)
    writer.writeheader()
    for row in table_data:
        if row:
            writer.writerow(row)
```
